```python
import numpy as np
def conv_forward(X,num_classes,num_hidden_nodes,maxpooling1=False,maxpooling2=False):
    #####CONVOLUTION1#### 
    #(H_prev,W_prev)=X.shape 
    filter1=[] 
    num_f1=6 
    np.random.seed(0) 
    for f in range(6):
        filter1.append(np.random.randint(-1,5,25).reshape(5,5))
    f1=np.asarray(filter1) 
    #print("f1=",f1)
    print("filter_shape=",f1.shape)
    #print("f1=",f1)
    #"""
    #for f in range(6):
    #    f1.append(np.random.randint(2,size=(2,2)))
    
    H1=(X.shape[0])-(f1.shape[1])+1
    W1=(X.shape[1])-(f1.shape[2])+1
    print("H1=",H1)  
    print("W1=",W1)
    C1=np.zeros(num_f1*H1*W1).reshape(num_f1,H1,W1)
    b1=np.ones(H1*W1).reshape(H1,W1)
    print("B1_shape=",b1.shape)
    for n_f1 in range(num_f1):
        for h1 in range(H1):
            for w1 in range(W1):
                x1_slice=X[h1:h1+(f1.shape[1]),w1:w1+(f1.shape[2])]
                C1[n_f1,h1,w1]=ReLU(np.sum(x1_slice*f1[n_f1])+b1[h1][w1])
    final_C1=C1
    #print("\nfinal_C1=",final_C1)
    #final_C1=C1+b1
    print("final_C1_shape",final_C1.shape)
    #print("final_C1",final_C1)

    ####MAXPOOL1####

    if maxpooling1:
        pool1=np.ones(13*13).reshape(13,13)
        print("pool1_shape=",pool1.shape)
        HP1=(final_C1.shape[1])-(pool1.shape[0])+1
        WP1=(final_C1.shape[2])-(pool1.shape[1])+1
        print("HP1=",HP1)  
        print("WP1=",WP1)
        CP1=np.zeros(num_f1*HP1*WP1).reshape(num_f1,HP1,WP1)
        for n_f1 in range(num_f1):
            for hp1 in range(HP1):
                for wp1 in range(WP1):
                    p1_slice=final_C1[n_f1][hp1:hp1+(pool1.shape[0]), wp1:wp1+(pool1.shape[1])]
                    CP1[n_f1,hp1,wp1]=np.max(p1_slice*pool1)
    
    #print("CP1=",CP1)
    print("\nshape_after_pooling=",CP1.shape)
    #print("CP1=",CP1)   
    #return CP1
    
    #####CONVOLUTION2#### 
    filter2=[] 
    num_f2=12
    np.random.seed(1) 
    for f in range(12):
        filter2.append(np.random.randint(-1,5,25).reshape(5,5))
    f2=np.asarray(filter2) 
    #print("f2=",f2)
    print("\nfilter2_shape=",f2.shape)
    #print("f2=",f2)
    #"""
    H2=(CP1.shape[1])-(f2.shape[1])+1
    W2=(CP1.shape[2])-(f2.shape[2])+1
    print("H2=",H2)  
    print("W2=",W2)
    C2=np.zeros(num_f2*H2*W2).reshape(num_f2,H2,W2)
    b2=np.ones(H2*W2).reshape(H2,W2)
    print("B2_shape=",b2.shape)
    for n_f2 in range(12):
        for n_f1 in range(6):
            for h2 in range(H2):
                for w2 in range(W2):
                    x2_slice=CP1[n_f1][h2:h2+(f2.shape[1]),w2:w2+(f2.shape[2])]
                    #print(x2_slice.shape,n_f2)
                    C2[n_f2,h2,w2]=ReLU(np.sum(x2_slice*f2[n_f2])+b2[h2][w2])
    
    final_C2=C2
    print("final_C2_shape",final_C2.shape)
    
    ###MAXPOOLING2###
    
    if maxpooling2:
        pool2=np.ones(4*4).reshape(4,4)
        print("pool2_shape=",pool2.shape)
        HP2=(final_C2.shape[1])-(pool2.shape[0])+1
        WP2=(final_C2.shape[2])-(pool2.shape[1])+1
        print("HP2=",HP2)  
        print("WP2=",WP2)
        CP2=np.zeros(num_f2*HP2*WP2).reshape(num_f2,HP2,WP2)
        for n_f2 in range(num_f2):
            for hp2 in range(HP2):
                for wp2 in range(WP2):
                    p2_slice=final_C2[n_f2][hp2:hp2+(pool2.shape[0]), wp2:wp2+(pool2.shape[1])]
                    CP2[n_f2,hp2,wp2]=np.max(p2_slice*pool2)
    
    #print("CP2=",CP2)
    print("\nshape_after_pooling=",CP2.shape)
    #print("CP2=",CP2)   
    #return CP2
    
    ###FLATTENING###
    
    flat=[]
    for layer in range(CP2.shape[0]):
        for h_flat2 in range(CP2.shape[1]):
            for w_flat2 in range(CP2.shape[2]):
                flat.append(CP2[layer][h_flat2][w_flat2])
    flattened_layer=np.asarray(flat).reshape(12*5*5,1)
    print("flattened_layer=",flattened_layer)
    print("\nshape of flatted_layer",flattened_layer.shape)
    
    ###FULLY CONNECTED LAYERS###
    
    hidden_layer=np.zeros(num_hidden_nodes).reshape(num_hidden_nodes,1)
    output_layer=np.zeros(num_classes).reshape(num_classes,1)
    
    np.random.seed(3)
    WIH = np.random.normal(0, 1/np.sqrt(flattened_layer.shape[0]),(flattened_layer.shape[0],hidden_layer.shape[0])) 
    WHO = np.random.normal(0, 1/np.sqrt(hidden_layer.shape[0]),(hidden_layer.shape[0],output_layer.shape[0])) 
    #WIH=np.random.rand(-2,2,flattened_layer[0]*hidden_layer[0]).reshape(flattened_layer[0],hidden_layer[0])
    #WHO=np.random.randn(-0.2,0.2,hidden_layer[0]*output_layer[0]).reshape(hidden_layer[0],output_layer[0])
    b_hidden=np.ones(num_hidden_nodes).reshape(num_hidden_nodes,1)
    b_output=np.ones(num_classes).reshape(num_classes,1)
    print("WIH=",WIH)
    print("WHO=",WHO)
    print("WIH_shape=",WIH.shape)
    print("WHO_shape",WHO.shape)
    hidden_layer=np.dot(WIH.T,flattened_layer)+b_hidden
    print(hidden_layer.shape)
    activation_h=[]
    for node in range(hidden_layer.shape[0]):
        activation_h.append(ReLU(hidden_layer[node][0]))
    activation_hidden=np.asarray(activation_h).reshape(num_hidden_nodes,1)
    print("hidden_layer_shape=",activation_hidden.shape)
    output_layer=np.dot(WHO.T,activation_hidden)+b_output
    activation_o=[]
    for node in range(output_layer.shape[0]):
        activation_o.append(ReLU(output_layer[node][0]))
    activation_output=np.asarray(activation_o).reshape(num_classes,1)
    print("output_layer=",activation_output)
    print("output_layer_shape=",activation_output.shape)
    predicted_value=np.argmax(activation_output)
    print("predicted_value=",predicted_value)
```
